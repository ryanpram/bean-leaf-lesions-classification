## Transfer Learning

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications.xception import preprocess_input
```

```python
# setup train generator with prepared preprocessing function
train_gen = ImageDataGenerator(preprocessing_function=preprocess_input)
```

```python
# generate training data automatically with tensorflow function help
train_ds = train_gen.flow_from_directory(
    './data/train', # path to train dir
    target_size=(299,299), # intended target size for training image
    batch_size=32 # amount of image per batch
)
```

```python
#get next batch of train dataset
X, y = next(train_ds)
```

```python
# check train ds shape
X.shape
```

```python
# show first 5 target label with one hot encoding
y[:5]
```

```python
# generate val dataset
val_gen = ImageDataGenerator(preprocessing_function = preprocess_input)
val_ds = val_gen.flow_from_directory(
    './data/val',
    target_size=(299,299),
    batch_size=32,
    shuffle=False
)
```

```python
# get pretrained model from Xception, and we treat it as our base model
base_model = Xception(
    weights='imagenet',
    include_top=False, # only get base model and not include dense layer
    input_shape= (299,299,3)
)

# freeze the base model
# we only want to finetune the dense layer
base_model.trainable = False

#create input layer
inputs = keras.Input(shape=(299,299,3))

#apply input to base model
base = base_model(inputs, training=False)

#do pooling the output of applied base model with input, to vectorize output to 1 dimension
vectors = keras.layers.GlobalAveragePooling2D()(base)


#take the vector and applied with dense with 10 neuron as output
# 3 because we have 3 classes on our case
outputs = keras.layers.Dense(3)(vectors)

#create keras model with inputs and outputs
model = keras.Model(inputs, outputs)
```

```python
preds = model.predict(X)
```

```python
preds.shape
```

```python
# show pred of first image
# showing that the output is random number as we havent train the model
preds[0]
```

```python
learning_rate = 0.01
optimizer = keras.optimizers.Adam(learning_rate = learning_rate)

loss = keras.losses.CategoricalCrossentropy(from_logits=True)

model.compile(optimizer=optimizer, loss=loss, metrics=['accuracy'])
```

```python
history = model.fit(train_ds, epochs=10, validation_data=val_ds)
```

```python
plt.plot(history.history['accuracy'], label='train')
plt.plot(history.history['val_accuracy'], label='val')
plt.xticks(np.arange(10))
plt.legend()
```

## Adjusting the learning rate

```python
def make_model(learning_rate=0.01):
    base_model = Xception(
        weights='imagenet',
        include_top=False, 
        input_shape= (299,299,3)
    )

    base_model.trainable = False
    
    ########################################################

    inputs = keras.Input(shape=(299,299,3))

    base = base_model(inputs, training=False)

    vectors = keras.layers.GlobalAveragePooling2D()(base)


    outputs = keras.layers.Dense(3)(vectors)

    model = keras.Model(inputs, outputs)
    
    ########################################################
    
    optimizer = keras.optimizers.Adam(learning_rate = learning_rate)

    loss = keras.losses.CategoricalCrossentropy(from_logits=True)

    model.compile(
        optimizer=optimizer, 
        loss=loss, 
        metrics=['accuracy'])
    
    return model
```

```python
scores = {}

for lr in [0.0001, 0.001,0.01,0.1]:
    print(lr)
    model = make_model(learning_rate=lr)
    history = model.fit(train_ds, epochs=10, validation_data=val_ds)
    scores[lr] = history.history
    
    print()
    print()
```

```python
for lr, hist in scores.items():
    plt.plot(hist['val_accuracy'], label=lr)

plt.xticks(np.arange(10))
plt.legend()
```

```python
# keep 2 best learning rate model
del scores[0.001]
del scores[0.0001]
```

```python
for lr, hist in scores.items():
    plt.plot(hist['accuracy'], label=('train=%s' % lr))
    plt.plot(hist['val_accuracy'], label=('val=%s' % lr))

plt.xticks(np.arange(10))
plt.legend()
```

From the initial experiement with learning rate adjument we got **0.01** as the learning rate to generate the best result

```python
best_learning_rate = 0.01
```

## Checkpointing

```python
checkpoint = keras.callbacks.ModelCheckpoint(
    './saved_model/xception_v1_{epoch:02d}_{val_accuracy:.3f}.h5', #filename template
    save_best_only=True,
    monitor='val_accuracy', #parameter of monitor
    mode='max', # find max val accuracy
)
```

```python
model = make_model(learning_rate=best_learning_rate)

history = model.fit(
    train_ds, 
    epochs=10, 
    validation_data=val_ds,
    callbacks=[checkpoint] # set additional paramater "callback"
)
```

## Adding More Layer

```python
def make_model(learning_rate=0.01, size_inner=100):
    base_model = Xception(
        weights='imagenet',
        include_top=False, 
        input_shape= (299,299,3)
    )

    base_model.trainable = False
    
    ########################################################

    inputs = keras.Input(shape=(299,299,3))

    base = base_model(inputs, training=False)

    vectors = keras.layers.GlobalAveragePooling2D()(base)
    
    #additional inner layer
    inner = keras.layers.Dense(size_inner, activation='relu')(vectors)


    outputs = keras.layers.Dense(3)(inner)

    model = keras.Model(inputs, outputs)
    
    ########################################################
    
    optimizer = keras.optimizers.Adam(learning_rate = learning_rate)

    loss = keras.losses.CategoricalCrossentropy(from_logits=True)

    model.compile(
        optimizer=optimizer, 
        loss=loss, 
        metrics=['accuracy'])
    
    return model
```

```python
scores = {}

for size in [10, 100, 1000]:
    print(size)
    
    model = make_model(learning_rate=best_learning_rate, size_inner=size)
    history = model.fit(train_ds, epochs=10, validation_data=val_ds)
    scores[size] = history.history
    
    print()
    print()
```

```python
for size, hist in scores.items():
    plt.plot(hist['val_accuracy'], label=('val=%s' % size))

plt.xticks(np.arange(10))
plt.legend()
```

From the experiement with additional inner layer  we got **100** as the best size of inner layer to generate the best result

```python
best_inner_size = 100
```

## 8.9 Regularization and dropout

```python
def make_model(learning_rate=0.01, size_inner=100, droprate=0.5):
    base_model = Xception(
        weights='imagenet',
        include_top=False, 
        input_shape= (299,299,3)
    )

    base_model.trainable = False
    
    ########################################################

    inputs = keras.Input(shape=(299,299,3))
    base = base_model(inputs, training=False)
    vectors = keras.layers.GlobalAveragePooling2D()(base)
    
    #additional inner layer
    inner = keras.layers.Dense(size_inner, activation='relu')(vectors)
    drop = keras.layers.Dropout(droprate)(inner)

    outputs = keras.layers.Dense(3)(inner)

    model = keras.Model(inputs, outputs)
    
    ########################################################
    
    optimizer = keras.optimizers.Adam(learning_rate = learning_rate)
    loss = keras.losses.CategoricalCrossentropy(from_logits=True)

    model.compile(
        optimizer=optimizer, 
        loss=loss, 
        metrics=['accuracy'])
    
    return model
```

```python
learning_rate = 0.001
size = 100

scores = {}

for droprate in [0.0, 0.2, 0.5, 0.8]:
    print(droprate)

    model = make_model(
        learning_rate=best_learning_rate,
        size_inner=best_inner_size,
        droprate=droprate
    )

    history = model.fit(train_ds, epochs=30, validation_data=val_ds)
    scores[droprate] = history.history

    print()
    print()
```

#### Comparison graph of diff drop rate

```python
for droprate, hist in scores.items():
    plt.plot(hist['val_accuracy'], label=('val=%s' % droprate))

plt.ylim(0.82, 0.98)
plt.legend()
```

#### Compare val accuracy graph between without regularization and with regularization (w best droprate)

```python
hist = scores[0.0]
plt.plot(hist['val_accuracy'], label=0.0)

hist = scores[0.8]
plt.plot(hist['val_accuracy'], label=0.8)

plt.legend()
```

we can see that with regularization (dropout), we can get better model result in validation setting 

#### Check graph of val and training accuracy of droprate 0.8

```python
hist = scores[0.8]
plt.plot(hist['accuracy'], label='training')

hist = scores[0.8]
plt.plot(hist['val_accuracy'], label='validation')

plt.legend()
```

the above graph is showing in droprate **0.8**, we got quite stable result without overfitting indication 

```python
best_droprate = 0.8
```

## Final Model

Here sum up all experiment, with train final model with all best params

```python
def make_model(input_size=150,learning_rate=0.01, size_inner=100, droprate=0.5):
    base_model = Xception(
        weights='imagenet',
        include_top=False, 
        input_shape= (input_size,input_size,3)
    )

    base_model.trainable = False
    
    ########################################################

    inputs = keras.Input(shape=(input_size,input_size,3))
    base = base_model(inputs, training=False)
    vectors = keras.layers.GlobalAveragePooling2D()(base)
    
    #additional inner layer
    inner = keras.layers.Dense(size_inner, activation='relu')(vectors)
    drop = keras.layers.Dropout(droprate)(inner)

    outputs = keras.layers.Dense(3)(inner)

    model = keras.Model(inputs, outputs)
    
    ########################################################
    
    optimizer = keras.optimizers.Adam(learning_rate = learning_rate)
    loss = keras.losses.CategoricalCrossentropy(from_logits=True)

    model.compile(
        optimizer=optimizer, 
        loss=loss, 
        metrics=['accuracy'])
    
    return model
```

```python
input_size = 299
```

```python
train_gen = ImageDataGenerator(
    preprocessing_function=preprocess_input,
)

train_ds = train_gen.flow_from_directory(
    './data/train',
    target_size=(input_size, input_size),
    batch_size=32
)


val_gen = ImageDataGenerator(preprocessing_function=preprocess_input)

val_ds = train_gen.flow_from_directory(
    './data/val',
    target_size=(input_size, input_size),
    batch_size=32,
    shuffle=False
)
```

```python
checkpoint = keras.callbacks.ModelCheckpoint(
    './saved_model/xception_bean_leaf_{epoch:02d}_{val_accuracy:.3f}.h5',
    save_best_only=True,
    monitor='val_accuracy',
    mode='max'
)
```

```python
model = make_model(
    input_size=input_size,
    learning_rate=best_learning_rate,
    size_inner=best_inner_size,
    droprate=best_droprate
)

history = model.fit(train_ds, epochs=50, validation_data=val_ds,
                   callbacks=[checkpoint])
```
